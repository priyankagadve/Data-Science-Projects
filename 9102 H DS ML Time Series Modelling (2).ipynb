```python
# Time Series Patterns :
# Time Series patterns are long term increase / decrease in observations.

# Type of Patterns : 

# 1. Seasonal Patterns :
# When certain period affects long term increase / decrease in observations.

# 2. Cyclic Patterns :
# 

# 3. Irregular Patterns :
# When long term increase / decrease can not be defined
```

```python
# Time Series Modelling :

# 1. Stationary Data      :       To build Time Series Model Stationary Data is required.
# 2. Non Stationary Data  :       To build Time Series Model Non stationary data needs to be converted into stationary data.      
```

```python
# Example : Time Series Modelling 

# Dataset : US Airline Dataset

# Objective :
# 1. Check data is Stationary.
# 2. If not => Convert Non Stationary Data into Stationary Data .
```

```python
# Import Libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import dateutil.parser as parse
```

```python
df = pd.read_csv("AirPassengers.csv", parse_dates= ['Month'], index_col= ['Month'])
```

```python
df.tail()
```

Stationary / Non Stationary :

Rolling Mean :

Example 

Obs.    Rolling Mean (Window = 3)
10            -----
15            -----
20            (10 + 15 + 20) / 3 = 15
25            (25 + 20 + 15) / 3 = 20 
30                               = 25
35                               = 30
40                               = 35


Observations :

=> If Rolling Mean is increasing then the data is 'Non Stationary'
=> If Rolling Mean values are confined within a certain range then the data is stationary.

```python
# Plot Passangers Vs Manth
```

```python
plt.figure(figsize = (10,6))
plt.plot(df['#Passengers'])
```

```python
# Since average at any interval is always greater than the previous value => Data is Non - Stationary.

# If Data is Non Stationary, It needs to converted into Stationary Data.

# Method : 

# 1. Decomposition 
#          A. Additive Decomposition
#          B. Multiplicative Decomposition

# 2. Differencing

# 3. Transformations :
#          (Square root , Cube root , Log, etc. )
```

```python
# 1. Additive Decomposition : 
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
res_add = seasonal_decompose(df['#Passengers'], model= 'additive', period = 1 )
```

```python
# plot Additive Decomposition 
```

```python

res_add.plot()
plt.show()
```

```python
# Multiplicative Decomposition 
```

```python
res_mult = seasonal_decompose(df['#Passengers'], model= 'multiplicative', period = 1 )
```

```python
res_mult.plot()
```

```python
# Differencing 
```

```python
# delta = Y(t) - Y(t-1)
```

```python
df['#Passengers_diff'] = df['#Passengers'] - df['#Passengers'].shift(1)
```

```python
df
```

```python
df['#Passengers_diff'].dropna().plot()
```

```python
# Transformations 
```

```python
# Y(t)   == > Transform   ==>   Process Data   ==>  Inverse Transform   ==>  Y(t)
```

```python
# Square root, cube root, log 
```

```python
# 1. Log Transformtion 

df['log'] = np.log(df['#Passengers'])

# 2. Square Root Transformation 

df['sqrt'] = np.sqrt(df['#Passengers'])

# 3. Cube Root Transformation

df['cube root'] = np.cbrt(df['#Passengers'])
```

```python

df['log'].plot()
```
